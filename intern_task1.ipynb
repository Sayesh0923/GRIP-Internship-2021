```python

# Step 1 :- Importing all libraries required in this notebook

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline

```

```python

# Step 2 :- Reading data from external source


s_data = pd.read_csv("http://bit.ly/w-data")
print("Data imported successfully")

# To check whether data is correctly imported so check first 10 rows

s_data.head(10)
```

```python
#Since we do have two variables lets plot simple scatter plot to check what our dataset says  

# Step 3 :- Plotting the distribution of scores vs no of hours studied

s_data.plot(x='Hours', y='Scores', style='o' , color='red')  
plt.title('Hours vs Scores')  
plt.xlabel('Hours Studied by student')  
plt.ylabel('Scores gained by student')  
plt.show()

```

```python
# lets prepare data for evalvation

# Step 4 :- Divide the data into attributes(inputs) and labels(outputs) for train and test case


X = s_data.iloc[:, :1].values  
y = s_data.iloc[:, -1].values


#Now that we have our attributes and labels, the next step is to split this data into training and test sets.
#We do this by using Scikit-Learn's built-in train_test_split() method


from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=0)
```

```python
# Step 5 :- Training the Algorithm

from sklearn.linear_model import LinearRegression  
refer = LinearRegression()  
refer.fit(X_train, y_train) 

print("Training complete")
```

```python
# Step 6 :- To visualize Plot the regression line

line = refer.coef_*X+refer.intercept_

plt.scatter(X, y,c='r')
plt.plot(X, line);
plt.xlabel('Hours Studied by student')  
plt.ylabel('Scores gained by student')
plt.show()
```

```python

# Step 7 :- Since Now that we have trained our algorithm, it's time to make some predictions.

# Testing data - In Hours
print(X_test) 

# Predicting the scores
y_pred = refer.predict(X_test) 

# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```python
# Step 8 :- Time to predict any random data

hrs = [9.25]
new_pred = refer.predict([hrs])
print("No of Hours = {}".format(hrs))
print("Predicted Score of student after studying for {} hours = {}".format(hrs,new_pred[0]))
```

```python

# Steo 9 :- Always have a check of Mean Absolute Error , To evaluate the performance of algorithm 

from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
